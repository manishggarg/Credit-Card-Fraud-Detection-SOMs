# Credit Card Fraud Detection:

We use real data set with real credit card transactions that had been anonymize for us.

```python
# To handle datasets
import pandas as pd
import numpy as np

# for plotting
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# to display all the columns of the dataframe in the notebook
pd.pandas.set_option('display.max_columns', None)
```

```python
# Import data 
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
# rows and columns of the data
print(df.shape)
```

```python
# Feature Scaling 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df['normalizedAmount'] = sc.fit_transform(df['Amount'].values.reshape(-1,1))
df = df.drop(['Amount'],axis=1)
```

```python
df.head()
```

```python
df = df.drop(['Time'],axis=1)
df.head()
```

```python
# split into features ans Labels
X = df.drop(['Class'],axis=1)
y = df['Class']

#Same as above
# X = data.iloc[:, data.columns != 'Class']
# y = data.iloc[:, data.columns == 'Class']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Deep neural network:

```python
# Importing the Keras libraries and packages 
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model = Sequential([
    Dense(units=16, input_dim = 29,activation='relu'),
    Dense(units=24,activation='relu'),
    Dropout(0.5),
    Dense(20,activation='relu'),
    Dense(24,activation='relu'),
    Dense(1,activation='sigmoid'),
])
```

```python
model.summary()
```

```python
# Compiling the Neural Network
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Fitting the Neural Network to the Training set
model.fit(X_train,y_train,batch_size=15,epochs=6)
```

```python
score = model.evaluate(X_test, y_test)
```

```python
print(score)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cnf_matrix = confusion_matrix(y_test, y_pred.round())
cnf_matrix
```

## Sampling:
Oversampling and undersampling in data analysis are techniques used to adjust the class distribution of a data set (i.e. the ratio between the different classes/categories represented). These terms are used both in statistical sampling, survey design methodology and in machine learning.
### UnderSampling:

```python
# choose the index of sample which  have class 1(Fraud)
fraud_indices = np.array(df[df.Class == 1].index)
```

```python
type(df[df.Class == 1].index)
```

```python
number_records_fraud = len(fraud_indices)
print(number_records_fraud)
```

```python
# choose the index of sample which  have class 0(normal)
normal_indices = df[df.Class == 0].index
normal_indices
```

```python
# choose random samples(no.=492)
random_normal_indices= np.random.choice(normal_indices, number_records_fraud, replace=False)
```

```python
random_normal_indices
```

```python
print(len(random_normal_indices))
```

```python
under_sample_indices = np.concatenate([fraud_indices,random_normal_indices])
print(len(under_sample_indices))
```

```python
under_sample_data = df.iloc[under_sample_indices,:]
under_sample_data.head()
```

```python
# split into features ans Labels
under_X = under_sample_data.drop(['Class'],axis=1)
under_y= under_sample_data['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(under_X,under_y, test_size=0.3)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,batch_size=15,epochs=5)
```

```python
y_pred = model.predict(X_test)
cnf_matrix = confusion_matrix(y_test, y_pred.round())
cnf_matrix
```
